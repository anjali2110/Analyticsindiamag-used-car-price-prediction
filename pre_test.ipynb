```python
import pandas as pd
import numpy as np

df= pd.read_excel('Data_Test.xlsx')
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df=df.drop('New_Price', 1)
df.head()
```

```python
df=df.dropna()
df.isna().sum()
```

```python
print(df.Location.unique())
print()
print(df.Year.unique())
print()
print(df.Fuel_Type.unique())
print()
print(df.Transmission.unique())
print()
print(df.Owner_Type.unique())
print()
print(df.Seats.unique())
```

```python
# for  name 
new = df["Name"].str.split(" ", n = 1, expand = True) 
df["Name"]= new[0] 
df.head()

df['Name']= df['Name'].str.lower()
print(df.Name.unique())
```

```python
# for mileage name 
new = df["Mileage"].str.split(" ", n = 1, expand = True) 
df["Mileage"]= new[0] 
df.head()
```

```python
# for engine name 
new = df["Engine"].str.split(" ", n = 1, expand = True) 
df["Engine"]= new[0] 
df.head()
```

```python
# for power name 
new = df["Power"].str.split(" ", n = 1, expand = True) 
df["Power"]= new[0] 
df.head()
```

```python
#remove null values from Price

df=df[df['Power'] != 'null']
df.head()
```

```python
#remove null values from Price

df= df[df['Mileage'] != '0']
df.head()
```

```python
#remove null values in Seats
df= df[df['Seats'] != 0]
df.head()
```

```python
categorical_cols=['Name','Location','Year','Fuel_Type','Transmission','Owner_Type','Seats']
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df[categorical_cols] = df[categorical_cols].apply(lambda col: le.fit_transform(col))
df[categorical_cols].head(10)
```

```python
df.head()
```

```python
df.to_csv('lb_test_data.csv',index=False)
```

```python
df['new_name'] = pd.get_dummies(df.Name).values.tolist()
df['new_location'] = pd.get_dummies(df.Location).values.tolist()
df['new_year'] = pd.get_dummies(df.Year).values.tolist()
df['new_fuel_type'] = pd.get_dummies(df.Fuel_Type).values.tolist()
df['new_transmission'] = pd.get_dummies(df.Transmission).values.tolist()
df['new_seats'] = pd.get_dummies(df.Seats).values.tolist()

df.head()
```

```python
train_cols=['Kilometers_Driven','Mileage','Engine','Power','new_seats','new_name','new_location','new_year','new_fuel_type','new_transmission','Price']
train = df[train_cols]
train.head()
```
